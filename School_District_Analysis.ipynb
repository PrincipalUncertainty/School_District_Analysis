```python
import pandas as pd
```

```python
# Import the data from the new_full_student_data.csv file into a 
# DataFrame named student_df by using the Pandas read_csv function 
# and the os module.

student_data_csv_path = "new_full_student_data.csv"

student_data_df = pd.read_csv(student_data_csv_path)
student_data_df

# Confirm that Pandas correctly imported the data by using 
# the head function
student_data_df.head()
```

```python
# In the student DataFrame, check for rows that have NaN 
# (or missing) values, and remove those rows

df2 = student_data_df.dropna()

df2.head()
```

```python
df2.isna().sum().to_frame()
```

```python
df2 = df2.drop_duplicates()

df2.head()
```

```python
# In the student DataFrame, check for rows that have NaN 
# (or missing) values, and remove those rows

df2 = df2.dropna()

df2.head()
```

```python
# In the student DataFrame, check for duplicate rows, and remove them.

df2.duplicated().to_frame()
```

```python
# Check the data types 

df2.dtypes.to_frame()
```

```python
# In the grade column, remove the "th" suffix from every value 
# by using str and replace,

df2.loc[:, "grade"] = df2.loc[:, "grade"].str.replace("th", "")
df2
```

```python
# Confirm removal of "th" suffix

df2.loc[:, 'grade'].to_frame()
```

```python
# Change the "grade" column to the int type, 
# and then verify the column types

df2['grade'] = df2['grade'].astype('int')
df2.dtypes
```

```python
# Store the row with the minimum overall reading score in min_reading_row 
# by using loc and the min_reading_score variable

min_reading_score = df2['reading_score'].min()
min_reading_score
min_reading_row = df2.loc[df2['reading_score'] == min_reading_score]
min_reading_row
```

```python
# Select all the reading scores from the 10th graders 
# at Dixon High School by using loc with conditionals

reading_scores_grade_10 = df2.loc[(df2["grade"] == 10) & (df2["school_name"] == "Dixon High School")]
reading_scores_grade_10 = reading_scores_grade_10['reading_score']
reading_scores_grade_10.to_frame()
```

```python
# Find the mean reading score for all students in 
# grades 11 and 12 combined.

mean_reading_11_12 = df2.loc[(df2["grade"] == 11) | (df2["grade"] == 12), "reading_score"].mean()
mean_reading_11_12
```

```python
# school budget mean by school type / grade

avg_budget_pergrade_perschool = df2[["school_budget","school_type"]].groupby(by=['school_type'], as_index = True).mean()
avg_budget_pergrade_perschool
```

```python
# Use the `groupby`, `count`, and `sort_values` functions to find the
# total number of students at each school and sort from most students to least students.

df2[["student_id", "school_name"]].groupby(by=['school_name']).count().sort_values(by=["student_id"], ascending = False)

```

```python
# Calculate the average math score by grade for each school type

avg_budget_pergrade_perschool = df2[["math_score","school_type", "grade"]].groupby(by=['school_type', "grade"], as_index = True).mean()
avg_budget_pergrade_perschool
```
